```python
## 5. ë¬¸ì„œ ì²˜ë¦¬ ë° ë²¡í„° ì €ì¥
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=600, ## ì²­í¬ì˜ ìµœëŒ€ ê¸¸ì´
    chunk_overlap=200, ## ì²­í¬ ê°„ ì¤‘ë³µë˜ëŠ” ê¸¸ì´
    length_function=len, ## ê¸¸ì´ ì¸¡ì • í•¨ìˆ˜
    separators=["\n\n", "\n", " ", ""], ## ë¶„í• ì— ì‚¬ìš©í• 
    keep_separator=True, ## êµ¬ë¶„ìë¥¼ ì²­í¬ì— í¬í•¨í• ì§€ ì—¬ë¶€
    is_separator_regex=False ## êµ¬ë¶„ìê°€ ì •ê·œí‘œí˜„ì‹ì¸ì§€ ì—¬ë¶€
)

for file_path in all_files:
    print(f"Processing file: {file_path}")
    loader = PyMuPDFLoader(file_path)
    pdf_pages = loader.load()
    print(f"Number of pages in the PDF: {len(pdf_pages)}")

    all_chunks = []
    all_metadatas = []

    for i, pdf_page in enumerate(pdf_pages):
        pdf_text = pdf_page.page_content ## ê° í˜ì´ì§€ì˜ í…ìŠ¤íŠ¸
        pdf_metadata = pdf_page.metadata ## ê° í˜ì´ì§€ì˜ ë©”íƒ€ë°ì´í„°
        pdf_chunks = text_splitter.split_text(pdf_text) ## ê° í˜ì´ì§€ì˜ ì²­í¬ë“¤
        print(f"Page {i} has {len(pdf_chunks)} chunks.") 

        all_chunks.extend(pdf_chunks)
        all_metadatas.extend([pdf_metadata] * len(pdf_chunks))
    print(f"Total metadata entries from file {file_path}: {len(all_metadatas)}")
    print(f"Sample metadata: {all_metadatas[0] if all_metadatas else 'No metadata'}")
    print(f"Total chunks from file {file_path}: {len(all_chunks)}")

    # ì„ë² ë”© ìƒì„±
    embeddings = embedding_model.embed_documents(all_chunks) 
    print(f"Generated {len(embeddings)} embeddings.")

    # Milvusì— ë°ì´í„° ì‚½ì…
    data = [
        {'id': idx, 'vector': embeddings[idx], 'text': all_chunks[idx], 'meta': all_metadatas[idx]}
        for idx in range(len(embeddings))
    ]

    res = client.insert(collection_name='kmu_llm', data=data)
    print(f"Inserted {res['insert_count']} vectors into Milvus from file {file_path}.")

```
