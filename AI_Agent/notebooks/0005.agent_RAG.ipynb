# Agentic AI(AgentAI)란 무엇인가

## 개요
**Agentic AI**(줄여서 *AgentAI*라고도 부름)는 사용자가 설정한 **목표를 달성하기 위해 스스로 판단·계획·행동**하는 인공지능 에이전트 패러다임입니다. 전통적 모델이 “명령→응답”에 머무는 반면, Agentic AI는 **자율성(agency)**을 바탕으로 **작업을 분해하고, 외부 도구를 호출하며, 순환적으로 목표 달성**을 시도합니다.  [oai_citation:0‡Amazon Web Services, Inc.](https://aws.amazon.com/what-is/agentic-ai/?utm_source=chatgpt.com)

---

## 핵심 정의
- **자율적 목표 달성 시스템**: 사전에 정의된 목표를 향해 **인간의 지속적 개입 없이** 계획을 세우고 실행합니다.  [oai_citation:1‡Amazon Web Services, Inc.](https://aws.amazon.com/what-is/agentic-ai/?utm_source=chatgpt.com)  
- **에이전트(Agent)**: 환경을 인지하고(Perception) → 의사결정(Reasoning/Decision) → **행동(Action)**으로 이어지는 루프를 수행하는 소프트웨어 구성요소입니다. 필요 시 다른 에이전트와 협업합니다.  [oai_citation:2‡IBM](https://www.ibm.com/think/topics/ai-agents?utm_source=chatgpt.com)  
- **오케스트레이션(Orchestration)**: 복수 에이전트가 각 하위 과업을 맡아 **조정**되는 구조로, 실시간 문제 해결을 지향합니다.  [oai_citation:3‡IBM](https://www.ibm.com/think/topics/agentic-ai?utm_source=chatgpt.com)

## 구성 요소(참고 아키텍처)
1. **목표·제약**: 사용자가 최상위 목표와 안전 제약을 부여  
2. **계획 모듈**: 목표를 **작업·하위작업**으로 분해(Planning)  
3. **도구 사용기(툴콜러)**: 검색, RAG, API, 코드 실행 등 **외부 도구 호출**  
4. **메모리/지식 소스**: 벡터DB·그래프DB·업무시스템 등 상태 저장/조회  
5. **오케스트레이터**: **복수 에이전트 협업·순서 제어·예외 처리**  
6. **평가·검증 루프**: 결과를 **스스로 채점/비판(Self-critique)**하고 수정  
7. **거버넌스 레이어**: **권한·감사·콘텍스트 경계·보안** 정책 적용

```python
## Agentic RAG 파이프라인 구현
# 필요한 라이브러리 Import
from pymilvus import MilvusClient
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_community.tools.tavily_search import TavilySearchResults
from langgraph.graph import StateGraph, END
from langgraph.checkpoint.memory import MemorySaver
from dataclasses import dataclass, field
from typing import List, Optional, Any, Literal, Dict
from dotenv import load_dotenv
import os
import json
import logging

# 환경 변수 로드
load_dotenv()

# 로깅 설정
logging.basicConfig(level=logging.INFO)

# ── API 키 설정 및 검증 ──
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY", "")

if not OPENAI_API_KEY:
    raise RuntimeError("OPENAI_API_KEY 환경 변수가 필요합니다.")

# LLM 및 임베딩 모델 초기화
try:
    llm = ChatOpenAI(
        model="gpt-4o-mini", 
        temperature=0.1, 
        openai_api_key=OPENAI_API_KEY
    )
    embedding_model = OpenAIEmbeddings(
        model="text-embedding-3-small", 
        openai_api_key=OPENAI_API_KEY
    )
except Exception as e:
    raise RuntimeError(f"LLM 초기화 실패: {e}")

# Milvus 데이터베이스 연결
DB_DIR = '/Users/sonny/Desktop/llm_seminar/lecture2/AI_Agent/retrieve'
DB_NAME = 'kmu.db'
DB_URL = os.path.join(DB_DIR, DB_NAME)

try:
    client = MilvusClient(DB_URL)
except Exception as e:
    raise RuntimeError(f"Milvus 연결 실패: {e}")

# 웹검색 도구 초기화
web_search_tool = TavilySearchResults(max_results=5) if TAVILY_API_KEY else None
if not TAVILY_API_KEY:
    print("⚠️  TAVILY_API_KEY가 설정되지 않아 웹검색 기능이 비활성화됩니다.")

print("✅ 모든 도구가 성공적으로 초기화되었습니다.")

# ─────────────────────────────────────────────────────────────────────────────
# 1) 파이프라인 설정 및 상수
# ─────────────────────────────────────────────────────────────────────────────
# 데이터베이스 설정
COLLECTION_NAME = "kmu_llm"

# 검색 및 필터링 설정
MIN_LOCAL_HITS = 3            # 로컬 검색 결과 개수 임계치
MAX_TOP_DISTANCE = 0.35       # 코사인 거리 기준 (작을수록 유사)
SEARCH_LIMIT = 8              # 검색 결과 개수

# 텍스트 처리 설정
MAX_CHARS_PER_CHUNK = 1000    # 청크당 최대 문자 수
MAX_WEB_CONTENT_CHARS = 800   # 웹 컨텐츠 최대 문자 수
MAX_CONTEXT_CHARS = 2000      # 컨텍스트 최대 문자 수
COMPRESSED_CONTEXT_LIMIT = 800 # 압축된 컨텍스트 최대 문자 수
```

```python

# ─────────────────────────────────────────────────────────────────────────────
# 2) 유틸리티 함수
# ─────────────────────────────────────────────────────────────────────────────
def truncate(text: str, n: int = MAX_CHARS_PER_CHUNK) -> str:
    """문자열을 지정된 길이로 자르고 생략 표시를 추가합니다."""
    if not text:
        return ""
    return text if len(text) <= n else text[:n] + "..."

def hybrid_hits_to_list(hybrid_hits) -> List[Any]:
    """MilvusClient.search 반환 HybridHits → list[Hit]로 변환합니다."""
    if not hybrid_hits:
        return []
    return list(hybrid_hits[0]) if hybrid_hits else []

def build_context_from_hits(hits: List[Any]) -> Dict[str, Any]: 
    """
    Hit 객체 리스트에서 컨텍스트 문자열, 출처, top_distance를 생성합니다.
    
    Args:
        hits: Hit 객체 리스트 (.entity(dict), .distance(float) 속성 포함)
    
    Returns:
        Dict: context, citations, top_distance를 포함한 딕셔너리
    """
    if not hits:
        return {"context": "", "citations": [], "top_distance": None}

    parts, citations = [], []
    top_distance = getattr(hits[0], "distance", None)

    for i, hit in enumerate(hits):
        try:
            ent = getattr(hit, "entity", {}) or {}
            text = ent.get("text", "") or ""
            meta = ent.get("meta", {}) or {}
            source = meta.get("source", f"doc_{i+1}")
            
            text = truncate(text, MAX_CHARS_PER_CHUNK)
            parts.append(f"{i+1}. ({source})\n{text}")
            citations.append(f"[{source}]")
        except Exception as e:
            logging.warning(f"Hit 처리 중 오류: {e}")
            continue

    return {
        "context": "\n\n".join(parts), 
        "citations": citations, 
        "top_distance": top_distance
    }

```

```python

# ─────────────────────────────────────────────────────────────────────────────
# 3) 상태 클래스 정의
# ─────────────────────────────────────────────────────────────────────────────
@dataclass
class AgentState:
    """Agent의 상태를 관리하는 데이터 클래스"""
    question: str                                    # 원본 질문
    rewritten: Optional[str] = None                  # 재작성된 질문
    local_hits: List[Any] = field(default_factory=list)  # 로컬 검색 결과
    top_distance: Optional[float] = None             # 최상위 결과의 거리
    use_web: bool = False                           # 웹검색 사용 여부
    web_snippets: List[str] = field(default_factory=list)  # 웹검색 결과
    context_compact: str = ""                       # 압축된 컨텍스트
    citations: List[str] = field(default_factory=list)     # 출처 목록
    answer_draft: str = ""                          # 초안 답변
    answer_final: str = ""                          # 최종 답변
    verify_note: str = ""                           # 검증 노트
```

```python

# ─────────────────────────────────────────────────────────────────────────────
# 4) 노드 함수 정의 - 프롬프트 템플릿 및 체인 구성
# ─────────────────────────────────────────────────────────────────────────────

# 4-1) 질문 재작성 노드
rewrite_prompt = ChatPromptTemplate.from_messages([
    ("system", 
     "당신은 정보 검색 질의 재작성 전문가입니다. "
     "사용자의 질문을 검색에 최적화된 명확하고 구체적인 문장으로 재작성하세요.\n"
     "재작성 원칙:\n"
     "- 핵심 키워드와 엔티티를 명확히 포함\n"
     "- 불필요한 수식어 제거\n"
     "- 검색 친화적인 용어 사용\n"
     "- 한 문장으로 간결하게 작성"),
    ("user", "원본 질문: {q}\n\n재작성된 질문:")
])
rewrite_chain = rewrite_prompt | llm | StrOutputParser()

def node_rewrite(state: AgentState) -> AgentState:
    """질문을 검색 친화적으로 재작성합니다."""
    try:
        state.rewritten = rewrite_chain.invoke({"q": state.question}).strip()
        logging.info(f"질문 재작성 완료: {state.rewritten}")
    except Exception as e:
        logging.error(f"질문 재작성 실패: {e}")
        state.rewritten = state.question.strip()
    return state

# 4-2) 로컬 검색 노드
def node_retrieve_local(state: AgentState) -> AgentState:
    """Milvus DB에서 로컬 검색을 수행합니다."""
    query = state.rewritten or state.question
    
    try:
        # 쿼리 임베딩
        q_vec = embedding_model.embed_query(query)
        
        # Milvus 검색
        res = client.search(
            collection_name=COLLECTION_NAME,
            data=[q_vec],
            limit=SEARCH_LIMIT,
            output_fields=["id", "text", "meta"]
        )
        
        hits = hybrid_hits_to_list(res)
        state.local_hits = hits
        state.top_distance = getattr(hits[0], "distance", None) if hits else None
        
        logging.info(f"로컬 검색 완료: {len(hits)}개 결과, top_distance: {state.top_distance}")
        
    except Exception as e:
        logging.error(f"로컬 검색 실패: {e}")
        state.local_hits = []
        state.top_distance = None
    
    return state

# 4-3) 분기 조건: 웹검색 필요 여부 판단
def should_use_web(state: AgentState) -> Literal["local", "web"]:
    """로컬 검색 결과의 품질을 평가하여 웹검색 필요 여부를 결정합니다."""
    
    # 웹검색 도구가 없으면 로컬만 사용
    if not web_search_tool:
        logging.info("웹검색 도구 없음 - 로컬 검색만 사용")
        return "local"
    
    # 검색 결과 개수 부족
    if len(state.local_hits) < MIN_LOCAL_HITS:
        logging.info(f"검색 결과 부족 ({len(state.local_hits)} < {MIN_LOCAL_HITS}) - 웹검색 사용")
        return "web"
    
    # 유사도 점수 부족
    if state.top_distance is not None and state.top_distance > MAX_TOP_DISTANCE:
        logging.info(f"유사도 점수 부족 ({state.top_distance} > {MAX_TOP_DISTANCE}) - 웹검색 사용")
        return "web"
    
    logging.info("로컬 검색 결과 충분 - 로컬 검색만 사용")
    return "local"

# 4-4) 웹검색 노드
def node_web_search(state: AgentState) -> AgentState:
    """Tavily를 사용하여 웹검색을 수행합니다."""
    if not web_search_tool:
        logging.warning("웹검색 도구가 설정되지 않았습니다.")
        state.web_snippets = []
        return state
    
    query = state.rewritten or state.question
    
    try:
        results = web_search_tool.invoke({"query": query})
        snippets = []
        
        for i, r in enumerate(results):
            title = r.get("title", "") or "웹 검색 결과"
            content = truncate(r.get("content", "") or "", MAX_WEB_CONTENT_CHARS)
            url = r.get("url", "") or ""
            snippets.append(f"{i+1}. [{title}] {content}\n(출처: {url})")
        
        state.web_snippets = snippets
        state.use_web = True
        logging.info(f"웹검색 완료: {len(snippets)}개 결과")
        
    except Exception as e:
        logging.error(f"웹검색 실패: {e}")
        state.web_snippets = []
        state.use_web = True  # 웹검색을 시도했음을 표시
    
    return state

# 4-5) 컨텍스트 압축 노드 (로컬/웹 공용)
compress_prompt = ChatPromptTemplate.from_messages([
    ("system", 
     "당신은 QA용 컨텍스트 압축 전문가입니다.\n"
     "주어진 자료에서 질문에 답하는 데 필요한 핵심 정보만 추출하여 "
     f"{COMPRESSED_CONTEXT_LIMIT}자 이내로 요약하세요.\n"
     "압축 원칙:\n"
     "- 사실 정보를 변경하지 말고 그대로 유지\n"
     "- 질문과 관련 없는 내용 제거\n"
     "- 중요한 숫자, 날짜, 이름 등은 정확히 보존\n"
     "- 간결하면서도 완전한 답변이 가능하도록 압축"),
    ("user", "질문: {q}\n\n자료:\n{ctx}\n\n압축된 컨텍스트:")
])
compress_chain = compress_prompt | llm | StrOutputParser()

def node_compress(state: AgentState) -> AgentState:
    """컨텍스트를 압축하고 출처를 정리합니다."""
    q = state.rewritten or state.question
    
    # 웹검색 결과 또는 로컬 검색 결과 처리
    if state.use_web:
        raw = "\n\n---\n\n".join(state.web_snippets)
        cites = [f"[web:{i+1}]" for i in range(len(state.web_snippets))]
    else:
        # 로컬 Hit → 컨텍스트 생성
        built = build_context_from_hits(state.local_hits[:5])
        raw = built["context"]
        cites = built["citations"]

    if not raw.strip():
        state.context_compact = ""
        state.citations = cites
        logging.warning("압축할 컨텍스트가 없습니다.")
        return state

    try:
        compact = compress_chain.invoke({"q": q, "ctx": raw}).strip()
        logging.info(f"컨텍스트 압축 완료: {len(raw)} → {len(compact)} 문자")
    except Exception as e:
        logging.error(f"컨텍스트 압축 실패: {e}")
        compact = truncate(raw, MAX_CONTEXT_CHARS)

    state.context_compact = compact
    state.citations = cites
    return state

# 4-6) 답변 생성 노드
answer_prompt = ChatPromptTemplate.from_messages([
    ("system", 
     "당신은 제공된 컨텍스트를 바탕으로 정확하고 유용한 답변을 제공하는 AI 어시스턴트입니다.\n"
     "답변 원칙:\n"
     "- 컨텍스트에 포함된 정보만을 사용하여 답변\n"
     "- 컨텍스트에 정보가 없으면 '제공된 문서에서는 해당 정보를 찾을 수 없습니다.'라고 명확히 표시\n"
     "- 한국어로 자연스럽고 이해하기 쉽게 작성\n"
     "- 구체적이고 도움이 되는 답변 제공\n"
     "- 추측이나 가정은 하지 않음"),
    ("user", "질문: {q}\n\n컨텍스트:\n{ctx}\n\n답변:")
])
answer_chain = answer_prompt | llm | StrOutputParser()

def node_answer(state: AgentState) -> AgentState:
    """컨텍스트를 바탕으로 답변을 생성합니다."""
    q = state.rewritten or state.question
    
    try:
        state.answer_draft = answer_chain.invoke({
            "q": q, 
            "ctx": state.context_compact
        }).strip()
        logging.info("답변 생성 완료")
    except Exception as e:
        logging.error(f"답변 생성 실패: {e}")
        state.answer_draft = "죄송합니다. 답변 생성 중 오류가 발생했습니다."
    
    return state

# 4-7) Self-RAG 검증 노드
verify_prompt = ChatPromptTemplate.from_messages([
    ("system", 
     "당신은 답변 품질 검증 전문가입니다. "
     "주어진 컨텍스트와 답변의 일치성을 검토하고 필요시 개선사항을 제안하세요.\n"
     "검증 기준:\n"
     "- 답변이 컨텍스트 내용과 일치하는가?\n"
     "- 답변이 질문에 적절히 대답하는가?\n"
     "- 사실 오류나 왜곡이 없는가?\n"
     "- 답변이 명확하고 이해하기 쉬운가?\n\n"
     "JSON 형식으로만 응답하세요:\n"
     '{{\"ok\": true|false, \"reason\": \"검증 결과 설명\", \"revised\": \"수정된 답변 또는 빈 문자열\"}}'),
    ("user", "컨텍스트:\n{ctx}\n\n답변 초안:\n{ans}\n\n검증 결과 (JSON):")
])
verify_chain = verify_prompt | llm | StrOutputParser()

def node_verify(state: AgentState) -> AgentState:
    """답변을 검증하고 필요시 수정합니다."""
    try:
        raw_response = verify_chain.invoke({
            "ctx": state.context_compact, 
            "ans": state.answer_draft
        })
        
        # JSON 파싱 시도
        data = json.loads(raw_response)
        ok = bool(data.get("ok", False))
        reason = data.get("reason", "")
        revised = (data.get("revised") or "").strip()
        
        # 검증 결과에 따라 최종 답변 결정
        if ok or not revised:
            state.answer_final = state.answer_draft
        else:
            state.answer_final = revised
            
        state.verify_note = f"검증결과: {'통과' if ok else '수정됨'} - {reason}"
        logging.info(f"답변 검증 완료: {'통과' if ok else '수정됨'}")
        
    except json.JSONDecodeError as e:
        logging.error(f"검증 응답 JSON 파싱 실패: {e}")
        state.answer_final = state.answer_draft
        state.verify_note = "검증 실패: JSON 파싱 오류"
    except Exception as e:
        logging.error(f"답변 검증 실패: {e}")
        state.answer_final = state.answer_draft
        state.verify_note = f"검증 실패: {str(e)}"
    
    return state
```

```python

# ─────────────────────────────────────────────────────────────────────────────
# 5) LangGraph 그래프 구성
# ─────────────────────────────────────────────────────────────────────────────
# StateGraph 생성
graph = StateGraph(AgentState)

# 노드 추가
graph.add_node("rewrite", node_rewrite)
graph.add_node("retrieve_local", node_retrieve_local)
graph.add_node("web_search", node_web_search)
graph.add_node("compress", node_compress)
graph.add_node("answer", node_answer)
graph.add_node("verify", node_verify)

# 시작점 설정
graph.set_entry_point("rewrite")

# 엣지 연결
graph.add_edge("rewrite", "retrieve_local")

# 조건부 엣지: 로컬 검색 결과에 따라 분기
graph.add_conditional_edges(
    "retrieve_local",
    should_use_web,
    {
        "local": "compress",    # 로컬 검색 결과가 충분한 경우
        "web": "web_search",    # 웹검색이 필요한 경우
    }
)

# 나머지 선형 엣지
graph.add_edge("web_search", "compress")
graph.add_edge("compress", "answer")
graph.add_edge("answer", "verify")
graph.add_edge("verify", END)

# 그래프 컴파일
try:
    app = graph.compile()
    logging.info("LangGraph 컴파일 완료")
except Exception as e:
    logging.error(f"LangGraph 컴파일 실패: {e}")
    raise  
```

```python

# ─────────────────────────────────────────────────────────────────────────────
# 6) 실행 함수 및 결과 표시 함수
# ─────────────────────────────────────────────────────────────────────────────
def run_agent(question: str) -> AgentState:
    """Agent를 실행하고 결과를 반환합니다."""
    if not question.strip():
        raise ValueError("질문이 비어있습니다.")
    
    try:
        # 초기 상태 생성
        init = AgentState(question=question.strip())
        
        # 그래프 실행
        result_dict = app.invoke(
            init, 
            config={"configurable": {"thread_id": "rag-session-1"}}
        )
        
        # 딕셔너리를 AgentState 객체로 변환
        result = AgentState(
            question=result_dict.get("question", question),
            rewritten=result_dict.get("rewritten"),
            local_hits=result_dict.get("local_hits", []),
            top_distance=result_dict.get("top_distance"),
            use_web=result_dict.get("use_web", False),
            web_snippets=result_dict.get("web_snippets", []),
            context_compact=result_dict.get("context_compact", ""),
            citations=result_dict.get("citations", []),
            answer_draft=result_dict.get("answer_draft", ""),
            answer_final=result_dict.get("answer_final", ""),
            verify_note=result_dict.get("verify_note", "")
        )
        
        logging.info("Agent 실행 완료")
        return result
        
    except Exception as e:
        logging.error(f"Agent 실행 실패: {e}")
        raise

def display_result(result: AgentState) -> None:
    """결과를 보기 좋게 출력합니다."""
    print("\n" + "="*80)
    print("📋 AGENTIC RAG 실행 결과")
    print("="*80)
    
    print(f"\n📝 [원본 질문]\n{result.question}")
    
    if result.rewritten and result.rewritten != result.question:
        print(f"\n✏️  [재작성된 질문]\n{result.rewritten}")
    
    print(f"\n📄 [압축된 컨텍스트]\n{result.context_compact}")
    
    print(f"\n✅ [최종 답변]\n{result.answer_final}")
    
    print(f"\n🔍 [검증 노트]\n{result.verify_note}")
    
    # 검색 통계 상세 정보
    print(f"\n📊 [검색 통계]")
    print(f"   - 로컬 검색 결과: {len(result.local_hits)}개")
    if result.top_distance is not None:
        print(f"   - 최고 유사도: {result.top_distance:.4f}")
    print(f"   - 웹검색 사용: {'예' if result.use_web else '아니오'}")
    
    # 로컬 검색 결과 상세 정보
    if result.local_hits:
        print(f"\n🔍 [로컬 검색 결과 상세]")
        for i, hit in enumerate(result.local_hits[:5]):  # 상위 5개만 표시
            try:
                ent = getattr(hit, "entity", {}) or {}
                meta = ent.get("meta", {}) or {}
                source = meta.get("source", f"문서_{i+1}")
                distance = getattr(hit, "distance", "N/A")
                text_preview = ent.get("text", "")[:100] + "..." if len(ent.get("text", "")) > 100 else ent.get("text", "")
                
                print(f"   {i+1}. 출처: {source}")
                print(f"      유사도: {distance:.4f}" if isinstance(distance, (int, float)) else f"      유사도: {distance}")
                print(f"      내용: {text_preview}")
                print()
            except Exception as e:
                print(f"   {i+1}. 검색 결과 표시 오류: {e}")
    
    # 웹검색 결과 상세 정보
    if result.use_web and result.web_snippets:
        print(f"\n🌐 [웹검색 결과 상세]")
        for i, snippet in enumerate(result.web_snippets):
            print(f"   {i+1}. {snippet}")
            print()
    
    # 출처 정보
    if result.citations:
        print(f"\n📚 [출처 목록]")
        if result.use_web:
            # 웹검색 결과의 출처
            for i, snippet in enumerate(result.web_snippets):
                lines = snippet.split('\n')
                title_line = lines[0] if lines else f"웹 검색 결과 {i+1}"
                source_line = lines[1] if len(lines) > 1 and lines[1].startswith('(출처:') else ""
                print(f"   [web:{i+1}] {title_line}")
                if source_line:
                    print(f"           {source_line}")
        else:
            # 로컬 검색 결과의 출처
            for i, citation in enumerate(result.citations):
                if i < len(result.local_hits):
                    try:
                        hit = result.local_hits[i]
                        ent = getattr(hit, "entity", {}) or {}
                        meta = ent.get("meta", {}) or {}
                        source = meta.get("source", citation)
                        print(f"   {citation} - {source}")
                    except:
                        print(f"   {citation}")
                else:
                    print(f"   {citation}")
    
    if result.use_web:
        print("\n💡 로컬 검색 결과가 부족하여 웹검색을 사용했습니다.")
    
    print("\n" + "="*80)

# 시스템 초기화 완료 메시지
print(f"\n✅ Agentic RAG 시스템 초기화 완료!")
print(f"📍 Milvus DB: {DB_URL}")
print(f"📚 Collection: {COLLECTION_NAME}")
print(f"🌐 웹검색: {'활성화' if web_search_tool else '비활성화'}")
print("\n🚀 사용법: run_agent('질문') 또는 display_result(run_agent('질문'))")
print("-" * 60)
```
